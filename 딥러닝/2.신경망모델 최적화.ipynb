```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#iris dataset load
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

# 신경망 구조생성
### 드롭아웃, 활성화함수 지정

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout, PReLU, Activation, ELU

model = Sequential()
model.add(Input(shape=(4,)))     #입력층:4개의 특징 (꽃받침 길이/너비, 꽃잎 길이/너비)
model.add(Dense(50, activation='sigmoid')) # 은닉층1: 뉴런 50개, sigmoid 활성화
model.add(Dense(50, activation=PReLU())) #은닉층2: 뉴런 50개, PReLU 활성화
model.add(Dropout(0.4))        # Dropout: 40% 뉴런 무작위로 꺼서 과적합 방지
model.add(Dense(30))            # 은닉층3: 뉴런 30개, 아직 활성화 없음
model.add(Activation(ELU(alpha=1)))    # ELU 활성화 (alpha=1)
model.add(Dense(3, activation='softmax'))  # 출력층: 클래스 3개 (Setosa, Versicolor, Virginica)
model.summary()
```

# 가중치 초기화

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout, PReLU, Activation, ELU
from tensorflow.keras.initializers import RandomNormal

model1 = Sequential()
model1.add(Input(shape=(4,)))
model1.add(Dense(50, activation='sigmoid', kernel_initializer='he_normal')) #가중치를 정규분포(평균 0, 분산 = 2/n)에 맞게 초기화
model1.add(Dense(50, activation=PReLU()))
model1.add(Dense(30, activation='sigmoid', kernel_initializer=RandomNormal(mean=0, stddev=0.1)))
model1.add(Dense(3, activation='softmax'))
model1.summary()
```

## 배치정규화
- 은닉층의 출력값 분포가 계속 바뀌는 문제가 발생할 때,
- 각 층의 출력을 평균=0, 표준편차=1로 정규화

```python
from tensorflow.keras.layers import BatchNormalization
model2 = Sequential()
model2.add(Input(shape=(4,)))
model2.add(Dense(50, activation='sigmoid')) #은닉층1: 뉴런50개
model2.add(BatchNormalization())            #은닉층1의 출력을 정규화
model2.add(Dense(50, activation='sigmoid'))
model2.add(BatchNormalization())
model2.add(Dense(30, activation='sigmoid'))
model2.add(BatchNormalization())
model2.add(Dense(3, activation='softmax'))
model2.summary()
```

# 컴파일

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy', #손실=정수레이블 기반 다중분류
    metrics=['accuracy']
)
model1.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy', #손실=정수레이블 기반 다중분류
    metrics=['accuracy']
)
model2.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy', #손실=정수레이블 기반 다중분류
    metrics=['accuracy']
)
```

# 학습

```python
model.fit(X_train, y_train, epochs=300, verbose=1)
#verbose=0 :아무것도 출력안함
#verbose=1 에포크마다 진행막대와 손실, 정확도 출력
#verbose=2 에포크별로 한 줄씩 간단 출력
```

```python

model1.fit(X_train, y_train, epochs=300, verbose=1)
```

```python
model2.fit(X_train, y_train, epochs=300, verbose=1)
```

# 평가

```python
model.evaluate(X_test, y_test)
model1.evaluate(X_test, y_test)
model2.evaluate(X_test, y_test)
```

## model, model1, model2 평가결과 비교

```python

# 공통 compile
for m in [model, model1, model2]:
    m.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 학습
history1 = model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0, validation_split=0.2)
history2 = model1.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0, validation_split=0.2)
history3 = model2.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0, validation_split=0.2)

# 평가
loss1, acc1 = model.evaluate(X_test, y_test, verbose=0)
loss2, acc2 = model1.evaluate(X_test, y_test, verbose=0)
loss3, acc3 = model2.evaluate(X_test, y_test, verbose=0)

print(" Model Evaluation Results")
print(f"Model   - Loss: {loss1:.4f}, Accuracy: {acc1:.4f}")
print(f"Model 1 - Loss: {loss2:.4f}, Accuracy: {acc2:.4f}")
print(f"Model 2 - Loss: {loss3:.4f}, Accuracy: {acc3:.4f}")
```

### 와인데이터 분류를 위한 딥러닝 모델

```python

import pandas as pd

redwine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=";")
X, y = redwine.iloc[:, :-1], redwine.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=1)

# Map the labels to be zero-indexed
unique_labels = np.unique(y_train) #품질 점수를 0부터 시작하는 정수로 매핑
label_map = {label: i for i, label in enumerate(unique_labels)}
y_train = y_train.map(label_map)
y_test = y_test.map(label_map) #와인 품질 점수 0부터 시작하는 연속적인 클래스 인덱스로 변환
```

```python
X.info()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.layers import PReLU, ELU, Activation, BatchNormalization
from tensorflow.keras.initializers import RandomNormal

model = Sequential()
model.add(Input(shape=(11,)))
model.add(Dense(100, activation='sigmoid', kernel_initializer=RandomNormal(mean=0, stddev= 0.1)))
model.add(BatchNormalization())
model.add(Dropout(0.2)) #뉴런 중 20% 랜덤으로 꺼서 과적합 방지.
model.add(Dense(150, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(100, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(50, activation='tanh', kernel_initializer='he_normal'))
model.add(BatchNormalization())
model.add(Dropout(0.1))
model.add(Dense(6, activation='softmax'))

#Dropout: 과적합 방지를 위한 노드 랜덤 비활성화
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=300, verbose=1, batch_size=200)
```

```python

model.evaluate(X_test, y_test)
```
